```python
import numpy as np
import pandas as pd

wine_red = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=';')

wine_red
```

```python
wine_red.info()
```

```python
X = wine_red.iloc[:, :-1] # 모든 행은 가져오고 마지막 열을 제외한 나머지 열을 x 에 저장
y = wine_red.iloc[:, -1]  # 모든 행은 가져오고 마지막 열만 y 에 저장
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()
LR.fit(x_train, y_train) # x_train 과 y_train 에 대해 학습시킴.
y_pred = LR.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

# evaluation MSE
mse = np.mean(np.square(y_pred-y_test))
mse_sk = mean_squared_error(y_pred, y_test)
print(f"mse: {mse}, mse_sk: {mse_sk}")

# evaluation MAE
mae = np.mean(np.abs(y_pred-y_test))
mae_sk = mean_absolute_error(y_pred, y_test)
print(f"mae: {mae}, mae_sk: {mae_sk}")

# evaluation r2_score
r2_sk = r2_score(y_pred, y_test)
print("r2_score:", r2_sk)
```

```python
from sklearn.datasets import fetch_california_housing

housing_dataset = fetch_california_housing()

print(housing_dataset.DESCR)
```

```python
# 열의 이름은 housing_dataset.feature_name 으로 하겠다.
df = pd.DataFrame(housing_dataset.data, columns=housing_dataset.feature_names)
df['target'] = housing_dataset.target

df
```

```python
X = housing_dataset.data
y = housing_dataset.target

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()
LR.fit(x_train, y_train)
y_pred = LR.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

# evaluation MSE
mse = np.mean(np.square(y_pred-y_test))
mse_sk = mean_squared_error(y_pred, y_test)
print(f"MSE: {mse}, MSE_SK: {mse_sk}")

# evaluation MAE
mae = np.mean(np.abs(y_pred-y_test))
mae_sk = mean_absolute_error(y_pred, y_test)
print(f"MAE: {mae}, MAE_SK: {mae_sk}")

# evaluation r2_score
r2_sk = r2_score(y_pred, y_test)
print("R2_SCORE:", r2_sk)
```
