## regression

```python
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()

X = housing.data
y = housing.target
```

```python
# 입력 데이터 하나가 8차원
print(X[0])
print(X[0].shape)
```

```python
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score

x_train, x_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# hidden layer 가 100차원까지 갈 수 있으므로 20 차원으로 설정, activation function 은 "relu" 로 설정
MLP_regr = MLPRegressor(hidden_layer_sizes=(20, ), activation='relu', max_iter=500)
MLP_regr.fit(x_train, y_train)
y_pred = MLP_regr.predict(x_test)

print(MLP_regr.loss_curve_)
print("r2 score:", r2_score(y_pred, y_test))
```

```python
import matplotlib.pyplot as plt

plt.plot(MLP_regr.loss_curve_)
```

```python
len(MLP_regr.loss_curve_)
```

## classification

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784')

X = mnist.data
y = mnist.target

print(X.shape, y.shape)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score

x_train, x_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

MLP_clf = MLPClassifier(hidden_layer_sizes=(100, ), activation='relu', max_iter=200)
MLP_clf.fit(x_train, y_train)
y_pred = MLP_clf.predict(x_test)

print(MLP_clf.loss_curve_)
print("f1 score:", f1_score(y_pred, y_test, average='micro'))
```

```python
import matplotlib.pyplot as plt

plt.plot(MLP_clf.loss_curve_)
```

```python
len(MLP_clf.loss_curve_)
```

# 과제

```python
# 당뇨 예측 데이터 사용
import pandas as pd

data = pd.read_csv("./data/diabetes_prediction_dataset.csv")
data.head(10)
```

```python
# 숫자가 아닌 column이 존재하기 때문에 해당 column은 encoding 진행
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
# encoder를 사용해서 각 column에 대해 fit_transform() 적용
data['gender'] = encoder.fit_transform(data['gender'])
data['smoking_history'] = encoder.fit_transform(data['smoking_history'])
```

```python
# X와 y로 나누기
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
# scikit-learn의 MLPClassifier를 사용하여 분류 진행
# 모델의 hidden_layer_sizes 및 다른 파라미터들은 적당한 값으로 자유롭게 조정
# 훈련 전에 데이터 스케일링 진행
# 모델 훈련 후 모델의 loss curve와 f1 score 출력하기
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score

x_train, x_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# hidden layer 개수를 늘린다고 성능이 좋아지는 것은 아님... (150, 70, 32, ) 은 0.966, (150, ) 은 0.973 정도의 정확도 나옴.
MLP_clf = MLPClassifier(hidden_layer_sizes=(150, 70, 32, ), activation='relu', max_iter=200)
MLP_clf.fit(x_train, y_train)
y_pred = MLP_clf.predict(x_test)

print(MLP_clf.loss_curve_)
print("f1 score:", f1_score(y_pred, y_test, average='micro'))
```

```python
# 모델의 loss curve를 그래프로 간단하게 그려보기 (matplotlib 사용)
import matplotlib.pyplot as plt

plt.plot(MLP_clf.loss_curve_)
```
