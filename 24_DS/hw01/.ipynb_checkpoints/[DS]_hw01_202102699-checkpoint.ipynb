### Mechanize 기본 사용법
아래에서 'data'에 해당 웹페이지의 소스 텍스트가 저장됨.

```python
import mechanize
import http.cookiejar

cj = http.cookiejar.CookieJar()
br = mechanize.Browser()
br.set_cookiejar(cj)

br.open("https://www.weather.go.kr/w/obs-climate/land/past-obs/obs-by-day.do")

data = br.response().read().decode('utf-8')
# Do something here with `data`
```

```python
## 정규표현식으로 최고기온, 최저기온을 뽑아온다 (x(최저기온), y(최고기온) 형식으로) - Test
import re

## 최고기온 추출 ##
pattern_max = r'<span>최고기온:([-+]?\d*\.\d+|\d+)' ## 숫자 catch 
d_max=re.findall(pattern_max, data)

## 최저기온 추출 ##
pattern_min = r'<span>최저기온:([-+]?\d*\.\d+|\d+)' ## 숫자 catch 
d_min=re.findall(pattern_min, data)

# print(d)
print(d_max)
print(d_min)

print(len(d_max))
print(len(d_min))

## 뽑아낸 최저기온과 최고기온을 튜플로 저장 ##
final_data = list(zip(d_min, d_max))
print(final_data)
```

## 문제 1 코드

```python
import re

final_data = [] ## 모은 데이터가 저장될 list ##
year = 2023 ## init year ##

while True:
    for month in range(1, 13):
        url = "https://www.weather.go.kr/w/obs-climate/land/past-obs/obs-by-day.do?stn=108&yy="+str(year)+"&mm="+str(month)+"&obs=1"
        print("waiting... ", url)
        br.open(url)
        data = br.response().read().decode("utf-8")

        ## 최고기온 추출 ##
        pattern_max = r'<span>최고기온:([-+]?\d*\.\d+|\d+)' ## 숫자 catch 
        d_max=re.findall(pattern_max, data)
        
        ## 최저기온 추출 ##
        pattern_min = r'<span>최저기온:([-+]?\d*\.\d+|\d+)' ## 숫자 catch 
        d_min=re.findall(pattern_min, data)
        
        ## 뽑아낸 최저기온과 최고기온을 튜플로 저장 ##
        data_list = list(zip(list(map(float, d_min)), list(map(float, d_max))))
        final_data = final_data + data_list
        print(final_data)

    if (len(final_data) >= 1000):
        break
    else:
        year = year - 1

print(len(final_data))
```

```python
## 총 수집된 data 개수 ##
print(len(final_data))

## display 5 sample datas ##
for i in range(0, 5):
    print(final_data[i])
```

## 문제 2 답변

- 문제 1에서 사용한 data 수집방법은 아래와 같다.
  1. 2023년 1월부터 12월까지의 data 에서 최저기온과 최고기온을 각각 뽑아온다.
  2. 각각 뽑아온 데이터를 데이터 포인트 형식에 맞게 (최저기온, 최고기온) 의 tuple 로 만들어 list 에 저장한다
  3. 이렇게 저장해온 최종 list 의 size 가 1,000 이 넘지 않는다면 즉, 1,000 개의 data 를 모으지 못했다면 연도를 1년 줄여 2022년 1월 ~ 12월의 data 를 수집하는 방식으로 진행한다.
  4. 모은 data 가 1,000개가 넘는다면 반복을 끝내고 수집을 종료한다
 
- 사용한 정규표현식은 다음과 같다. (r' ' 로 감싸 준 이유는 \를 이스케이프로 인식하지 않도록 하기 위함.)
  1. r'\<span>최고기온:([-+]?\d*\.\d+|\d+)'
      - <span>최고기온: 으로 시작하며 그 뒤에 나오는 숫자에 대해 matching 시키는 정규표현식
      - [-+]?\d*\.\d+ 또는 \d+ 와 매치되는 숫자를 가져오며 앞의 패턴은 부호가 존재할 수 있으며, 소숫점이 있는 수와 매치된다. 뒤의 패턴은 정수와 매치되는 패턴이다.
  2. r'\<span>최저기온:([-+]?\d*\.\d+|\d+)'
     - 최고기온과 동일한 패턴이지만, 최저기온 뒤에 나오는 수와 매치되는 패턴이다. 

```python
## convert data list to dataframe ##
import pandas as pd

col_name = ['min_temp', 'max_temp']
temp_data = pd.DataFrame(final_data, columns=col_name)
```

## 문제3 최고기온과 최저기온에 대해 각각 히스토그램 그리기

```python
import seaborn as sns
import matplotlib.pyplot as plt

## 히스토그램 그리기 - minimum temperature ##
plt.hist(temp_data['min_temp'], bins=20)

## labeling ##
plt.title('Histogram of minimum temperature')
plt.xlabel('temperature (℃)')
plt.ylabel('frequency')

plt.show()

## 히스토그램 그리기 - maximum temperature ##
plt.hist(temp_data['max_temp'], bins=20)

## labeling ##
plt.title('Histogram of maximum temperature')
plt.xlabel('temperature (℃)')
plt.ylabel('frequency')

plt.show()
```

## 문제4 Scatter plot 과 contour plot 그리기

```python
## draw scatter plot ##
plt.scatter(x=temp_data['min_temp'], y=temp_data['max_temp']) 

## labeling ##
plt.title('temperature scatter plot')
plt.xlabel('minimum temperature (℃)')
plt.ylabel('maximum temperature (℃)')

plt.show()
```

```python
## draw contour plot - kind : kde ##
sns.jointplot(data=temp_data, x='min_temp', y='max_temp', kind='kde', fill='True')
plt.show()
```

```python
## draw contour plot - kind :Hex ##
sns.jointplot(data=temp_data, x='min_temp', y='max_temp', kind='hex')
plt.show()
```
