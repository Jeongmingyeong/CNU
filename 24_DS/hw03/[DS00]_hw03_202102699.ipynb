## 문제 1 코드

```python
import re

final_data = [] ## 모은 데이터가 저장될 list ##
year = 2023 ## init year ##

while True:
    for month in range(1, 13):
        url = "https://www.weather.go.kr/w/obs-climate/land/past-obs/obs-by-day.do?stn=108&yy="+str(year)+"&mm="+str(month)+"&obs=1"
        print("waiting... ", url)
        br.open(url)
        data = br.response().read().decode("utf-8")

        ## 최고기온 추출 ##
        pattern_max = r'<span>최고기온:([-+]?\d*\.\d+|\d+)' ## 숫자 catch 
        d_max=re.findall(pattern_max, data)
        
        ## 최저기온 추출 ##
        pattern_min = r'<span>최저기온:([-+]?\d*\.\d+|\d+)' ## 숫자 catch 
        d_min=re.findall(pattern_min, data)
        
        ## 뽑아낸 최저기온과 최고기온을 튜플로 저장 ##
        data_list = list(zip(list(map(float, d_min)), list(map(float, d_max))))
        final_data = final_data + data_list
        print(final_data)

    if (len(final_data) >= 1000):
        break
    else:
        year = year - 1

print(len(final_data))
```

```python
## convert data list to dataframe ##
import pandas as pd

col_name = ['min_temp', 'max_temp']
temp_data = pd.DataFrame(final_data, columns=col_name)
```

```python
import math
## 문제 1에서 사용하는 함수 정의 ##

# calculate mean (data_list -> data_list_size -> mean_value)
def mean(data_list, length):
    sum_value = 0
    for v in data_list:
        sum_value = sum_value + v
    return sum_value / length

# calculate median (data_list -> data_list_size -> median_value)
def median(data_list, length):
    data_list.sort()
    return data_list[int(length / 2)]

# calculate variance (data_list -> data_list_size -> mean -> variance)
def variance(data_list, length, mean):
    sum_temp = 0
    for v in data_list:
        sum_temp = sum_temp + ((v - mean) ** 2)
    return sum_temp / (length - 1)

# calculate standard deviation (variance -> standard deviation(√variance))
def standard_deviation(variance):
    return math.sqrt(variance)

# calculate the first quartile (data_list -> data_list_size -> nth_quartile -> first_quartile)
# if q == 1 then return value is the first quartile
def quartile(data_list, data_list_size, q):
    idx = int(((data_list_size + 1)/4) * q)
    return data_list[idx]

# calculate IQR (first_quartile -> third_quartile -> IQR)
def IQR(first_q, third_q):
    return third_q - first_q
```

```python
## total_length ##
totalsize = len(final_data)
max_temp_data_list = list(temp_data['max_temp'])
min_temp_data_list = list(temp_data['min_temp'])

## mean, median, standard_deviation, the_first_quartile, the_third_quartile, IQR구하기 ##

## about 최고기온 ##
max_temp_mean = mean(max_temp_data_list, totalsize)
max_temp_median = median(max_temp_data_list, totalsize)
max_temp_variance = variance(max_temp_data_list, totalsize, max_temp_mean)
max_temp_st_dev = standard_deviation(max_temp_variance)
max_temp_first_q = quartile(max_temp_data_list, totalsize, 1)
max_temp_third_q = quartile(max_temp_data_list, totalsize, 3)
max_temp_IQR = IQR(max_temp_first_q, max_temp_third_q)

## about 최저기온 ##
min_temp_mean = mean(min_temp_data_list, totalsize)
min_temp_median = median(min_temp_data_list, totalsize)
min_temp_variance = variance(min_temp_data_list, totalsize, min_temp_mean)
min_temp_st_dev = standard_deviation(min_temp_variance)
min_temp_first_q = quartile(min_temp_data_list, totalsize, 1)
min_temp_third_q = quartile(min_temp_data_list, totalsize, 3)
min_temp_IQR = IQR(min_temp_first_q, min_temp_third_q)

## 구한 값 출력 ##
print("[mean]")
print(f'최고기온 : {max_temp_mean} 최저기온 : {min_temp_mean}\n')

print("[median]")
print(f'최고기온 : {max_temp_median} 최저기온 : {min_temp_median}\n')

print("[standard deviation]")
print(f'최고기온 : {max_temp_st_dev} 최저기온 : {min_temp_st_dev}\n')

print("[the first and third quartile ]")
print(f'(first) 최고기온 : {max_temp_first_q} 최저기온 : {min_temp_first_q}')
print(f'(third) 최고기온 : {max_temp_third_q} 최저기온 : {min_temp_third_q}\n')

print("[IQR]")
print(f'최고기온 : {max_temp_IQR} 최저기온 : {min_temp_IQR}')
```

## 문제2 코드

```python
## 문제2에서 사용하는 함수 정의 ##

# calculate Skewness (data_list -> data_list_size -> mean -> skewness)
def Skewness(data_list, data_list_size, mean):
    under = 0 # 분모
    over = 0  # 분자
    
    for i in range(data_list_size):
        under = under + ((data_list[i] - mean) ** 2)
        over = over + ((data_list[i] - mean) ** 3)
        
    under = under ** (3/2)
    over = (data_list_size ** (1/2)) * over
    
    return over / under

# calculate Kurtosis (data_list -> data_list_size -> mean -> Kurtosis)
def Kurtosis(data_list, data_list_size, mean):
    under = 0 # 분모
    over = 0  # 분자
    
    for i in range(data_list_size):
        under = under + ((data_list[i] - mean) ** 2)
        over = over + ((data_list[i] - mean) ** 4)
        
    under = under ** 2
    over = data_list_size * over
    
    return (over / under) - 3
```

```python
## total_length ##
totalsize = len(final_data)
max_temp_data_list = list(temp_data['max_temp'])
min_temp_data_list = list(temp_data['min_temp'])

## Skewness, Kurtosis 구하기 ##

## about 최고기온 ##
max_temp_skewness = Skewness(max_temp_data_list, totalsize, max_temp_mean)
max_temp_kurtosis = Kurtosis(max_temp_data_list, totalsize, max_temp_mean)

## about 최저기온 ##
min_temp_skewness = Skewness(min_temp_data_list, totalsize, min_temp_mean)
min_temp_kurtosis = Kurtosis(min_temp_data_list, totalsize, min_temp_mean)

## 구한 값 출력 ##
print("[skewness]")
print(f'최고기온 : {max_temp_skewness} 최저기온 : {min_temp_skewness}\n')

print("[kurtosis]")
print(f'최고기온 : {max_temp_kurtosis} 최저기온 : {min_temp_kurtosis}')
```

```python
## calculate Pearson correlation coefficient

## total_length ##
totalsize = len(final_data)
max_temp_data_list = list(temp_data['max_temp'])
min_temp_data_list = list(temp_data['min_temp'])

temp = 0 # ∑(AB)

for i in range(totalsize):
    temp = temp + (max_temp_data_list[i] * min_temp_data_list[i])
    
temp1 = totalsize * max_temp_mean * min_temp_mean # nA'B' (A', B' is mean)
under = (totalsize - 1) * max_temp_st_dev * min_temp_st_dev # (n-1)𝜎a𝜎b

pcc = (temp - temp1) / under
print(f'Pearson correlation coefficient : {pcc}')
```

## 문제4 코드

- Null hypothesis (H0) : 2012년의 최저기온 평균과 2023년 최저기온 평균은 동일하다.(변하지 않았다)
- Alternative hypothesis (Ha) : 2023년 최저기온 평균은 2012년의 최저기온 평균보다 올랐다. (one-sided test)

- 𝝁0 = 2012년 최저기온 평균

```python
## 2012년과 2023년의 최저기온 모으기 ##
import re

final_data = [] ## 모은 데이터가 저장될 list (idx0 = 2012년, idx1 : 2023년) ##
year = ('2012', '2023')

for y in year:
    data_list = []
    for month in range(1, 13):
        url = "https://www.weather.go.kr/w/obs-climate/land/past-obs/obs-by-day.do?stn=108&yy="+y+"&mm="+str(month)+"&obs=1"
        print("waiting... ", url)
        br.open(url)
        data = br.response().read().decode("utf-8")

        ## 최저기온 추출 ##
        pattern_min = r'<span>최저기온:([-+]?\d*\.\d+|\d+)' ## 숫자 catch 
        d_min=re.findall(pattern_min, data)

        data_list += list(map(float, d_min))
        
    final_data.append(data_list)
    print(final_data)


print(len(final_data))
```

```python
## 2012년, 2023년 최저기온 평균구하기 ##
mean2012 = mean(final_data[0], len(final_data[0]))
mean2023 = mean(final_data[1], len(final_data[1]))

## 2012년 최저기온 표준편차구하기 ##
var = variance(final_data[0], len(final_data[0]), mean2012)
st_dev = standard_deviation(var)

## Zstat 구하기 ##
SE = st_dev / (len(final_data[0]) ** (1/2))
zstat = (mean2023 - mean2012) / SE

print(f'z statistics : {zstat}')
```

```python
## convert zstat to p-value ##
from scipy import stats

# zstat 을 p-value 로 변환할 때 사용하는 함수는 1-cdf 이다.
# zstat이 𝝁0에서 떨어진 정도를 나타내주는데 cdf 함수에 적용시키면 zstat보다 작은 쪽 면적 크기를 구하게 된다.
# 하지만 우리가 관심있는 건 zstat 보다 큰쪽의 면적이고, 다시말하면 전체 면적 1에서 cdf(zstat) 크기를 뺀 값이다.
# 따라서 이 값을 계산해주는 함수로 stats.norm.sf 함수가 존재하며, 이 함수를 사용해 p-value를 구해 가설 검정을 진행하였다.

print(f'p-value : {stats.norm.sf(zstat)}')
```

### p-value 가 0.1 이하이므로 Null hypothesis(H0) 를 강하게 reject 한다.
### 그러므로 Alternative hypothesis (H1) 으로 결론 도출 => 2012년 최저기온 평균에 비해 2023년 최저기온 평균온도가 올랐다.
